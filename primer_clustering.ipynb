## Importar librerías y carga de datos de prueba

```python
from sklearn.datasets import make_blobs
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x,y = make_blobs(n_samples=100,
                 centers = 4,
                 n_features=2,
                 cluster_std=[1,1.5,2,2],
                 random_state=7
                 )
```

```python
df_blobs = pd.DataFrame(
    {
        'x1':x[:,0],
        'x2':x[:,1],
        'y':y
    }
)
```

```python
df_blobs
```

```python
def plot_2d_clusters(x,y,ax):
    y_uniques = pd.Series(y).unique()
    for i in y_uniques:
        x[y==i].plot(
             title = f'{len(y_uniques)} Clusters',
             kind = 'scatter',
             x = 'x1',
             y = 'x2',
             marker = f'${i}$',
             ax = ax
         )
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,7))
x,y = df_blobs[['x1','x2']], df_blobs['y']
plot_2d_clusters(x,y,ax)
```

## Aplicando clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=7)
```

```python
y_predict = kmeans.fit_predict(x)
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,9))
plot_2d_clusters(x,y,axs[0])
plot_2d_clusters(x,y_predict,axs[1])
axs[0].set_title(f'Actual {axs[0].get_title()}')
axs[1].set_title(f'Kmeans {axs[1].get_title()}')
plt.tight_layout()
plt.show()
```
